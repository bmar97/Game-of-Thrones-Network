### Job Filter

```python
def job_filter(file):
    
    # Create DataFrame
    df = pd.read_csv (file + '.csv')
    
    # Filter for key words
    # ['sr', 'senior', 'intern', 'vp', 'president', 'director', 'staff', 
    # 'head', 'manager', 'manage', 'phd', 'ui', 'lead', 'front', 'full']
    
    # df['Location'].str.contains('CA') |
    
    df01 = df[df['Location'].str.contains('CA') | df['Location'].str.contains('NY')]
    
    df02 = df01[(df01['Job Title'].str.contains('sr', case=False) == False) & (
        df01['Job Title'].str.contains('senior', case=False) == False) & (
        df01['Job Title'].str.contains('intern', case=False) == False) & (
        df01['Job Title'].str.contains('vp', case=False) == False) & (
        df01['Job Title'].str.contains('president', case=False) == False) & (
        df01['Job Title'].str.contains('director', case=False) == False) & (
        df01['Job Title'].str.contains('staff', case=False) == False) & (
        df01['Job Title'].str.contains('head', case=False) == False) & (
        df01['Job Title'].str.contains('manager', case=False) == False) & (
        df01['Job Title'].str.contains('manage', case=False) == False) & (
        df01['Job Title'].str.contains('phd', case=False) == False) & (
        df01['Job Title'].str.contains('ui', case=False) == False) & (
        df01['Job Title'].str.contains('lead', case=False) == False) & (
        df01['Job Title'].str.contains('front', case=False) == False) & (
        df01['Job Title'].str.contains('full', case=False) == False)]
    
    # Increase index values by 2 and drop Date/VC columns 
    
    filtered = df02.reset_index()
    filtered['index'] = filtered['index'] + 2
    filtered = filtered.drop(labels=['Date', 'VC'], axis=1)
     
    return filtered

job_filter('/Users/branndonmarion/Desktop/JB-DFs/Data/Data (2.17)')
```

```python
def data_instantiation(date): 

    # Save job filtered DFs as excel files
    
    data = job_filter('/Users/branndonmarion/Desktop/JB-DFs/Data/Data ' + '('+ date + ')').to_excel('Data (' + date + ').xlsx')
    swe = job_filter('/Users/branndonmarion/Desktop/JB-DFs/SWE/SWE ' + '('+ date + ')').to_excel('SWE (' + date + ').xlsx')
    finance = job_filter('/Users/branndonmarion/Desktop/JB-DFs/Finance/Finance ' + '('+ date + ')').to_excel('Finance (' + date + ').xlsx')
    pso = job_filter('/Users/branndonmarion/Desktop/JB-DFs/PSO/PSO ' + '('+ date + ')').to_excel('PSO (' + date + ').xlsx') 
    it = job_filter('/Users/branndonmarion/Desktop/JB-DFs/IT/IT ' + '('+ date + ')').to_excel('IT (' + date + ').xlsx')
    
    return None

data_instantiation(2.24)
```
